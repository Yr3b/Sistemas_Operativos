#### Ejercicio 04-02

Consigna: Desarrolla un programa en python que simule una reserva de asientos en un cine. El cine tiene varias salas y cada sala tiene un conjunto de asientos. Los usuarios pueden reservar uno o mas asientos simultaneamente, y el sistema debe garantizar que no se produzcan reservas dobles para el mismo asiento.

```python
import threading,random,time

class Cine:
    def __init__(self,id,salas):
        self._salas = salas # Array de objetos sala
        self._id = id

class Sala:
    def __init__(self,asientos,id):
        self.__asientos = asientos # Array de objetos asiento
        self.__lock = threading.Lock()

    def reservar_asiento(self,cantidad_a_reservar):
        self.__lock.acquire()
        time.sleep(1)
        try:
            index = 0
            asientos_disponibles = [asiento for asiento in self.__asientos if not asiento._reservado]
            if len(asientos_disponibles) < cantidad_a_reservar:
                print(f"No hay suficientes asientos disponibles. Solo quedan {len(asientos_disponibles)}.\n")
                return

            for asiento in asientos_disponibles:
                if index < cantidad_a_reservar:
                    asiento.reservar()
                    index += 1
                    print(f"Asiento reservado: {asiento._id}\n")
                    print(f"Cantidad reservados: {index}\n")
            time.sleep(1)  # SimulaciÃ³n de tiempo en la reserva
        finally:
            self.__lock.release()  # Liberamos el lock manualmente

class Asiento:
    def __init__(self,id):
        self._reservado = False
        self._id = id

    def reservar(self):
        self._reservado = True

class Usuario(threading.Thread):
    def __init__(self,id,numero_asientos_a_reservar,sala,cine:Cine):
        super().__init__()
        self.__id = id
        self.__naar = numero_asientos_a_reservar
        self.__sala = sala
        self.__cine = cine

    def run(self):
        sala = self.__cine._salas[self.__sala]
        print(f"El usuario {self.__id} intenta reservar {self.__naar} asientos de la sala {self.__sala} del cine {self.__cine._id}\n")
        sala.reservar_asiento(self.__naar)

def main():
    asientos = []
    for i in range(20):
        asientos.append(Asiento(i))
    salas = []
    for j in range(5):
        salas.append(Sala(asientos,j))
    cine = Cine(1,salas)

    usuarios = []

    for i in range(5):
        usuarios.append(Usuario(i+1,i+1,0,cine))

    for usuario in usuarios:
        usuario.start()

    for usuario in usuarios:
        usuario.join()
main()

```
