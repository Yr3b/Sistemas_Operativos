#### Problema de los productores y consumidores (Problema del buffer bounded)

Consigna: Este es uno de los problemas clasicos de sincronizacion. Involuctra dos tipos de procesos, productores y consumidores, que comparten un buffer comun de tama√±o fijo. Los productores generan datos y los colocan en el buffer, mientras que los consumidores retiran datos de este buffer. El desafio es asegurar que los produtores no intenten agregar datos si el buffer esta lleno y que los consumidores no intenten retirar datos si el buffer esta vacio.

```python
import threading,time,random

class Productor(threading.Thread):
    def __init__(self,buffer,semaphore_producer,semaphore_consumer,buffer_lock):
        super().__init__()
        self._buffer = buffer
        self._semaphore_producer = semaphore_producer
        self._semaphore_consumer = semaphore_consumer
        self._buffer_lock = buffer_lock

    def run(self):
        for i in range(10):
            self._semaphore_producer.acquire()  # -1
            time.sleep(random.uniform(0.5, 1.5))
            with self._buffer_lock:
                self._buffer.append(f"Mensaje {i}")
                print(f"Produccion: {self._buffer}\n")
            self._semaphore_consumer.release()  # +1

                

class Consumidor(threading.Thread):
    def __init__(self,buffer,semaphore_producer,semaphore_consumer,buffer_lock):
        super().__init__()
        self._buffer = buffer
        self._semaphore_producer = semaphore_producer
        self._semaphore_consumer = semaphore_consumer
        self._buffer_lock = buffer_lock

    def run(self):
        for i in range(10):
            self._semaphore_consumer.acquire() # -1
            time.sleep(random.uniform(0.5, 1.5))
            with self._buffer_lock:
                consumo = self._buffer.pop()
                print(f"Consumidor consumio: {consumo}\n")
                print(f"Buffer: {self._buffer}\n")
            self._semaphore_producer.release() # +1


def main():
    buffer = []
    semaphore_producer = threading.Semaphore(4)
    semaphore_consumer = threading.Semaphore(0)
    buffer_lock = threading.Lock()
    consumidor = Consumidor(buffer,semaphore_producer,semaphore_consumer,buffer_lock)
    productor = Productor(buffer,semaphore_producer,semaphore_consumer,buffer_lock)

    consumidor.start()
    productor.start()

    consumidor.join()
    productor.join()
main()
```
