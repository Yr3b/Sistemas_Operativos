#### Ejercicio 06-01

Consigna: Implementar un sistema productor-consumidor en python utilizando hilos. En este sistema, un productor generara numeros enteros aleatorios y los colocara en una lista compartida. El consumidor debera consumir esos numeros (es decir, leerlos y eliminarlos de la lista) en el orden que fueron producidos. Usa un semaforo para controlar que el consumidor no intente consumir un numero cuando la lista este vacia.

```python
import threading,random,time

class Productor(threading.Thread):
    def __init__(self,listaCompartida,semaforo,max_items):
        super().__init__()
        self._listaCompartida = listaCompartida
        self._semaforo = semaforo
        self._max_items = max_items

    def generar_numero_aleatorio(self):
        return random.randint(0,100)

    def agregar_numero_aleatorio(self,numero_aleatorio):
        if(len(self._listaCompartida)<10):
            self._listaCompartida.append(numero_aleatorio)

    def run(self):
        for i in range(self._max_items):
            time.sleep(random.uniform(0.2,0.9))
            numero_aleatorio = self.generar_numero_aleatorio()
            print("\n Productor \n")
            print(f"Productor produjo: {numero_aleatorio}\n")
            self.agregar_numero_aleatorio(numero_aleatorio)
            print(f"Numero agregado ({numero_aleatorio}): {self._listaCompartida}\n")
            self._semaforo.release() # incrementa el semaforo

class Consumidor(threading.Thread): 
    def __init__(self,listaCompartida,semaforo,max_items):
        super().__init__()
        self._listaCompartida = listaCompartida
        self._semaforo = semaforo
        self._max_items = max_items

    def eliminar(self):
        itemEliminado = self._listaCompartida.pop()
        return itemEliminado

    def run(self):
        for i in range(self._max_items):
            self._semaforo.acquire()    # decrementa el semaforo, si esta en cero, espera...
            print("\n Consumidor \n")
            print(f"Lista antes de eliminar: {self._listaCompartida}\n")
            print(f"item Eliminado: {self.eliminar()}\n")
            print(f"Lista despues de eliminar: {self._listaCompartida}\n")
            time.sleep(random.uniform(0.5,1.5))
            
           

def main():
    listaCompartida = []
    semaforo = threading.Semaphore(0)
    max_items = 10
    productor = Productor(listaCompartida,semaforo,max_items)
    consumidor = Consumidor(listaCompartida,semaforo,max_items)

    productor.start()
    consumidor.start()

    productor.join()
    consumidor.join()
main()
```
